**Google Search Analysis using pytrends API**

**Google Trends provides API for analysing daily searches on Google. This API is called pytrends.**

```python
!pip install pytrends
```

```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq(hl='en-US', tz=360)
```

```python
trends.build_payload(kw_list=["Taylor Swift"])
data = trends.interest_by_region()
data = data.sort_values(by="Taylor Swift", ascending=False)
data.head()
```

```python
data.reset_index().plot(x="geoName", y="Taylor Swift", figsize=(100, 30), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=["Taylor Swift"])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(100, 30))
data["Taylor Swift"].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Taylor Swift', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total count')
plt.show()
```
